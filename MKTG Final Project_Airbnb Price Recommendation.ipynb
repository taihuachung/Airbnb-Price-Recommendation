#### Creating amenity variables

```python
df['TV'] = df['amenities'].str.contains('TV') * 1.0
df['Wifi'] = df['amenities'].str.contains('Wifi') * 1.0
df['Internet'] = df['amenities'].str.contains('Internet') * 1.0
df['Pets'] = df['amenities'].str.contains('Pet') * 1.0
df['Beach'] = df['amenities'].str.contains('Beach') * 1.0
df['Pool'] = df['amenities'].str.contains('Pool') * 1.0
df['Kitchen'] = df['amenities'].str.contains('Kitchen') * 1.0
df['Washer'] = df['amenities'].str.contains('Washer') * 1.0
df['Dryer'] = df['amenities'].str.contains('Dryer') * 1.0
df['Family_kid_friendly'] = df['amenities'].str.contains('Family/kid friendly') * 1.0
df['Free_street_parking'] = df['amenities'].str.contains('Free street parking') * 1.0
df['Hair_dryer'] = df['amenities'].str.contains('Hair dryer') * 1.0
df['Heating'] = df['amenities'].str.contains('Heating') * 1.0
```
